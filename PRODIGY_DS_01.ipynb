### Problem Statement

Create a bar chart or histogram to visualize the distribution of a
categorical or continuous variable, such as the distribution of
ages or genders in a population.

Sample Dataset:
https://github.com/Prodigy-InfoTech/data-science-datasets/tree/main/Task%201


### Importing Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(
    style="white",
    context="talk",      # larger, presentation-ready text
    font="sans-serif"
)
```

### Loading the Dataset

```python
df = pd.read_csv(
    "API_SP.POP.TOTL_DS2_en_csv_v2_7.csv",
    skiprows=4
)

df.head()
```

### Dataset Information

```python
df.info()
```

### Selecting Population Data for a Specific Year

```python
population_2020 = df[["Country Name", "2020"]].dropna()
population_2020.head()
```

### Bar Chart: Population Distribution (Top 10 Countries)

```python
top10 = population_2020.sort_values("2020", ascending=False).head(10)

plt.figure(figsize=(12,6))

bar = sns.barplot(
    data=top10,
    x="2020",
    y="Country Name",
    hue="Country Name",     
    palette="crest",
    legend=False            
)


plt.title(
    "Top 10 Most Populated Countries (2020)",
    fontsize=18,
    fontweight="bold",
    pad=15
)

plt.xlabel("Population", fontsize=14)
plt.ylabel("")

# Remove spines for modern look
sns.despine(left=True, bottom=True)

# Add value labels
for container in bar.containers:
    bar.bar_label(container, fmt='%.0f', padding=5)

plt.show()
```

### Histogram: Population Distribution

```python
# Set up the visual style
sns.set(style="whitegrid")

# Histogram for population distribution in 2020
plt.figure(figsize = (12, 6))
sns.histplot(data = population_2020, x='2020', bins=30, kde=True, color='blue')
plt.title('Population Distribution in 2020')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

### Observations

- The bar chart shows that a few countries have extremely high populations
  compared to others.
- The histogram indicates that most countries have relatively lower
  population values, while only a small number have very high populations.
- Population distribution is highly skewed.


### Conclusion

In this task, we visualized population data using both a bar chart and
a histogram. Bar charts helped compare population among countries,
